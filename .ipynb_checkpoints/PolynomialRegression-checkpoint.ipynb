```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

### 데이터 블러오기

```python
boston_df = pd.read_csv('boston.csv')
boston_df
```

### X 데이터 만들기

```python
X = boston_df.drop(columns="MEDV")
X
```

```python
X.columns
```

```python
polynomial_transformer = PolynomialFeatures(2)
```

```python
polynomial_data = polynomial_transformer.fit_transform(X)
```

```python
polynomial_data
```

```python
polynomial_features_names = polynomial_transformer.get_feature_names_out(X.columns)
```

```python
polynomial_features_names 
```

```python
X = pd.DataFrame(polynomial_data, columns=polynomial_features_names)
```

```python
X
```

### Y 데이터 만들기

```python
y = boston_df[["MEDV"]]
```

```python
y
```

### train, test 나누고 모델 학습

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
### θ₁
model.coef_
```

### 오차 확인으로 성능확인

```python
y_test_predict = model.predict(X_test)
```

```python
y_test_predict
```

```python
mse = mean_squared_error(y_test_predict, y_test)
mse = mse ** 0.5
mse
```
