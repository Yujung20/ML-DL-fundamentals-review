```python
import pandas as pd

titanic_df=pd.read_csv('titanic.csv')
titanic_df.head()
```

### 모듈 이용해서 원핫인코딩하기

- 원하는 열만 추출해서 원핫인코딩

```python
onehot_df=titanic_df[['Sex','Embarked']]
onehot_df
```

```python
one_hot_encoded_df=pd.get_dummies(onehot_df)
one_hot_encoded_df.head()
```

### 전체 프레임에서 원하는 데이터열만 원핫인코딩하기

```python
titanic_onehot_df = pd.get_dummies(data=titanic_df, columns=['Sex','Embarked'])
titanic_onehot_df.head()
```
