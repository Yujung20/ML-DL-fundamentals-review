```python
import torch
import torch.nn as nn
```

| index | 0 | 1 | 2 |
| --- | --- | --- | --- |
| 레이어 타입 |  Linear | Linear | Linear |
| 입력 차원 | 8 | 4 | 6 |
| 출력 차원 | 4 | 6 | 3 |
| 활성화 함수 | ReLU | ReLU | - |

```python
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear0 = nn.Linear(8, 4)
        self.linear1 = nn.Linear(4, 6)
        self.linear2 = nn.Linear(6, 3)
        #nn.Linear(in_feature, out_feature)의미
        self.relu = nn.ReLU()

    def forward(self, x):
        # h0_0 = self.linear0(x)
        # h0_1 = self.relu(h0_0)

        # h1_0 = self.linear1(h0_1)
        # h1_1 = self.relu(h1_0)

        # y = self.linear2(h1_1)

        x = self.relu(self.linear0(x))
        x = self.relu(self.linear1(x))
        y = self.linear2(x)

        return y
```

```python
model = MyModel()

input_tensor = torch.randn(2, 8)
result = model(input_tensor)
result

'''배치가 2여서 2행으로 나오고 피처 개수는 마지막 레이어(6,3)의 출력 차원에 맞춰서 3'''
```

```python
class MyModel2(nn.Module):
    def __init__(self):
        super().__init__()
        self.block = MyModel()
        self.layer = nn.Linear(3, 4)

    def forward(self, x):
        x = self.block(x)
        y = self.layer(x)
        return y
```

```python
model2 = MyModel2()

input_tensor = torch.randn(2, 8)
result = model2(input_tensor)
result
```
