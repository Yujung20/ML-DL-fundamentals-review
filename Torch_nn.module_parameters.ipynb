```python
import torch
import torch.nn as nn
```

```python
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear0 = nn.Linear(8, 4)
        self.linear1 = nn.Linear(4, 6)
        self.linear2 = nn.Linear(6, 3)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.linear0(x))
        x = self.relu(self.linear1(x))
        y = self.linear2(x)

        return y

model = MyModel()
```

```python
print(model)
```

```python
list(model.children())
```

```python
for child in model.children():
    print(child)
```

```python
for named_child in model.named_children():
    print(named_child)
```

```python
for param in model.parameters():
    print(param)
    print()
```

```python
for named_param in model.named_parameters():
    print(named_param)
    print()
```

```python
class MyModel2(nn.Module):
    def __init__(self):
        super().__init__()
        self.block = MyModel()
        self.linear = nn.Linear(3, 4)

    def forward(self, x):
        x = self.block(x)
        return self.linear(x)

model2 = MyModel2()
```

```python
print(model2)
```

```python
for named_child in model2.named_children():
    print(named_child)
```

```python
for named_param in model2.named_parameters():
    print(named_param)
    print()
```
