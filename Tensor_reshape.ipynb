```python
x = torch.randn(2, 3, 4)
permuted = x.permute(2, 0, 1)
'''permute는 차원의 인덱스를 입력
2가 0번 자리에 들어갔으니까, 2번 차원이 0번 차원이 됨'''
```

```python
permuted.size()
```

```python
x = torch.randn(1, 28, 1, 28)
```

```python
result0 = x.squeeze()
result0.size()
'''squeeze 메서도 사용하면 크기가 1인 차원 사라지는 경우'''
```

```python
result1 = x.squeeze(2)
result1.size()
'''인덱스가 2번 = 3번째 차원이 사라짐'''
```
