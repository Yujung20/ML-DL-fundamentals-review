```python
import numpy as np
from PIL import Image

class FlowerDataset(Dataset):
    def __init__(self):
        self.image_paths = glob.glob('./flower_photos/*/*.jpg')

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, index):
        image_path = self.image_paths[index]
        image = Image.open(image_path)
        image_np = np.array(image)
        return torch.tensor(image_np)
```

```python
import matplotlib.pyplot as plt

def show_image(image_tensor):
    image = image_tensor.numpy()
    plt.figure()
    plt.imshow(image)
    plt.axis('off')
```

```python
dataset = FlowerDataset()
```

```python
for i in range(5):
    show_image(dataset[i])
```

#### 04. Dataset 만들기 실습

```python
import numpy as np
from sklearn import datasets

import torch
from torch.utils.data import Dataset

cal_housing = datasets.fetch_california_housing()
input_data = cal_housing.data.astype(np.float32)
target_data = cal_housing.target.astype(np.float32)

class CaliforniaHousingDataset(Dataset):
    def __init__ (self, input_data, target_data):
        self.input_data=input_data
        self.target_data = target_data

    def __len__ (self):
        return len(self.input_data)

    def __getitem__(self, index) :
        input_tensor = torch.tensor(self.input_data[0])
        target_tensor = torch.tensor(self.target_data[0])
        return input_tensor, target_tensor


# 테스트 코드
dataset = CaliforniaHousingDataset(input_data, target_data)
print(dataset[0])
```
